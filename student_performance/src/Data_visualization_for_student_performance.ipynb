```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title('Math Scores by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Math Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='reading score', y='writing score', data=df, hue='gender')
plt.title('Reading vs Writing Scores')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='writing score', kde=True)
plt.title('Writing Score Distribution')
plt.xlabel('Writing Score')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='race/ethnicity', y='math score', data=df)
plt.title('Math Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Math Score')
plt.show()
```
